```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

plt.rcParams['figure.figsize'] = (16, 9)

# Creating a sample dataset with 4 clusters
X, y = make_blobs(n_samples=800, n_features=3, centers=4)
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X[:, 0], X[:, 1], X[:, 2])

# Initializing KMeans
kmeans = KMeans(n_clusters=4)
# Fitting with inputs
kmeans = kmeans.fit(X)
# Predicting the clusters
labels = kmeans.predict(X)
# Getting the cluster centers
C = kmeans.cluster_centers_

fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=y)
ax.scatter(C[:, 0], C[:, 1], C[:, 2], marker='*', c='#050505', s=1000)
```

```python
# import libraries
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
import json
```

```python
# import data

# fileName = './sample.json'
fileName = './../data/DMT_tags_vectors.json'

data = []
with open(fileName) as f:
    for line in f:
        data.append(json.loads(line))
```

```python
data[0]["vector"]
```

```python
# get vectors of each data
vectors = [];
pageIds = [];
pageTitles = [];
for page in data:
    vectors.append(page["vector"])
    pageIds.append(page["pageId"])
    pageTitles.append(page["pageTitle"])
```

```python
# convert to np arrays
np_vectors = np.array(vectors)
np_pageIds = np.array(pageIds)
np_pageTitles = np.array(pageTitles)
```

```python
print(np_vectors.shape)
print(np_pageIds.shape)
print(np_pageTitles.shape)
```

```python
# model
numClusters = 4
kmeans = KMeans(n_clusters=numClusters)
kmeans = kmeans.fit(np_vectors)
labels = kmeans.predict(np_vectors)
```
