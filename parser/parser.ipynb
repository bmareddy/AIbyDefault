```python
import json
```

```python
# fileName = './sample.json'
fileName = './../crawler/data/confluence/2018-06-09T21-39-16.json'
data = json.load(open(fileName))
```

```python
data[0]
```

```python
ancestorMapping = {};
for datum in data:
    currentPageId = datum['pageId']
    ancestorPages = datum['ancestors']
    if (len(ancestorPages) > 1):
        ancestorDict = {}
        ancestorDict['grandparentPageId'] = ancestorPages[1]['id']
        ancestorDict['grandparentPageTitle'] = ancestorPages[1]['title']
        
        if(len(ancestorPages) > 2):
            ancestorDict['parentPageId'] = ancestorPages[2]['id']
            ancestorDict['parentPageTitle'] = ancestorPages[2]['title']
            
        ancestorMapping[currentPageId] = ancestorDict
        
    
ancestorMapping
```
